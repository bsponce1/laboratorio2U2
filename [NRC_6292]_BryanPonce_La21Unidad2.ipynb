```python
'''Libreria para manipular los datos'''
import pandas as pd
'''Importación de librerías'''
import numpy as np
'''Librería para crear matrices o vectores'''
from sklearn.impute import SimpleImputer
'''Librería para sustituir valores nulos por otros valores '''
from sklearn.impute import KNNImputer
```

```python
'''cargar el dataset y separarlo por espacio'''
df=pd.read_csv("data_act_01.csv", sep=';')
'''Mostrar los 5 primeros datos'''
df.head()
```

# Valores nulos

```python
'''Dectectar datos nulos dentro del dataframe'''
df.isnull().sum()
```

# Eliminar columna innecesario

```python
'''Elimina la columna Range'''
df= df.drop(columns=['Range'], errors='ignore')
'''Elimina la columna CallDateTime'''
df= df.drop(columns=['CallDateTime'], errors='ignore')
'''Muestra valores sin la columnas'''
df
```

# Imputación de la columna City

```python
'''Sustituye los valores nulos por otros de la frequencia'''
imputer_mf = SimpleImputer(strategy='most_frequent', missing_values=np.nan)
'''Se entrena los valores de la columna City'''
imputer_mf= imputer_mf.fit(df[['City']])
'''Se transforma los valores por otros a la otra columna'''
df['City']=imputer_mf.transform(df[['City']])
'''Se analiza si el dataset esta limpio '''
df['City'].isnull().sum()
'''Se muestra el dataset'''
df
```

# Valores duplicados

```python
'''Verificar los valores duplicados en CrimeId'''
df.CrimeId.duplicated().sum()
```

```python
'''Se analiza el valor máximo para seguir la secuencia'''
df.CrimeId.max()
```

```python
'''Corrección de hora y fecha de tiempo de llamada'''
df['CallTime'] = pd.to_datetime(df['CallTime'])
'''Corrección de fecha de delito'''
df['OffenseDate'] = pd.to_datetime(df['OffenseDate'])
'''Transforma la columna los datos en formato hora'''
df['CallTime'].dt.hour
'''Transforma la columna los datos en dia de la semana'''
df['OffenseDate'].dt.dayofweek
'''Transforma la columna los datos en dia de la semana'''
'''Muestro el dataset'''
df
```

```python
'''Se elimina el id de crimen duplicado'''
df.drop_duplicates(['CrimeId'],keep='last')
```

```python
'''Verificacion de valores unicos'''
df['Disposition'].unique()
```

```python
'''Correccion de valores unicos mediante la moda'''
moda = df['Disposition'].mode()
'''Se remplaza el dato 22 por la moda de la columna'''
df['Disposition'] = df['Disposition'].replace('22',moda[0])
'''Se remplaza el dato Not por la moda de la columna'''
df['Disposition'] = df['Disposition'].replace('Not recorded',moda[0])
```

```python
'''Verificacion del cambio en dataset'''
df['Disposition'].unique()
```

```python
'''Verificacion de valores unicos en ciudad del dataset'''
df['City'].unique()
```

```python
'''Moda de la columna City'''
moda= df['City'].mode()
'''Muestro la moda'''
moda
```

```python
'''Correccion de SAN FRANCISCO por la moda la position del valor'''
df['City'] = df['City'].replace('SAN FRANCISCO',moda[0])
'''Correcion de nan por la moda'''
df['City'] = df['City'].replace(np.nan,moda[0])
'''Correcion de S por la moda'''
df['City'] = df['City'].replace(' S',moda[0])
```

```python
'''Verificacion si hay valores sucios'''
df['City'].unique()
```

```python
'''Verificar los valores este correctamente'''
df['State'].unique()
```

```python
'''Reemplazo en estado el valor nulo por el valor CA'''
df['State'] = df['State'].replace(np.nan,'CA')
```

```python
'''Verificar los valores este correctamente con los cambios'''
df['State'].unique()
```

```python
'''Verificar los valores este correctamente en la id agencia'''
df['AgencyId'].unique()
```

```python
'''Reemplazamos el CA por 1''
df['AgencyId'] = df['AgencyId'].replace('CA','1')
```

```python
'''Verificar los valores este correctamente con los cambios'''
df['AgencyId'].unique()
```
